- Job id: não se aplica

- auxiliaries qubits: 1

```python
executions_counts = [
    {'000': 22, '011': 242, '110': 247, '101': 217, '001': 230, '100': 34, '010': 32}
]
```

```python
# Plot results with legend

legend = []

for index in range(len(executions_counts)):
    legend.append("Execução: " + str(index + 1))
```
